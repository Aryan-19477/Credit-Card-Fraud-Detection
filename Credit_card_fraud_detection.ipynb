2.Data Preprocessing


```python
x=df.drop('Class',axis=1)
y=df['Class']
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.2,random_state=42)
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)
x_train, y_train = smote.fit_resample(x_train, y_train)
```

3.Model Development



```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(random_state=42)
lr.fit(x_train,y_train)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt= DecisionTreeClassifier(random_state=42)
dt.fit(x_train,y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier(random_state=42)
rf.fit(x_train,y_train)
```

4.Model Evaluation


```python
y_pred_lr=lr.predict(x_test)
y_pred_dt=dt.predict(x_test)
y_pred_rf=rf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

print("Accuracy:", round(accuracy_score(y_test, y_pred_lr)*100,3))
print("Precision:",round(precision_score(y_test, y_pred_lr)*100,3))
print("Recall:",round(recall_score(y_test, y_pred_lr)*100,3))
print("F1 Score:", round(f1_score(y_test, y_pred_lr)*100,3))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_lr))
```

```python
print("Accuracy:", round(accuracy_score(y_test, y_pred_dt)*100,3))
print("Precision:", round(precision_score(y_test, y_pred_dt)*100,3))
print("Recall:", round(recall_score(y_test, y_pred_dt)*100,3))
print("F1 Score:", round(f1_score(y_test, y_pred_dt)*100,3))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_dt))
```

```python
print("Accuracy:", round(accuracy_score(y_test, y_pred_rf)*100,3))
print("Precision:", round(precision_score(y_test, y_pred_rf)*100,3))
print("Recall:", round(recall_score(y_test, y_pred_rf)*100,3))
print("F1 Score:", round(f1_score(y_test, y_pred_rf)*100,3))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_rf))
```

```python

import pandas as pd

results = pd.DataFrame({
    'Model': ['Logistic Regression', 'Decision Tree', 'Random Forest'],
    'Accuracy': [
        round(accuracy_score(y_test, y_pred_lr) * 100, 3),
        round(accuracy_score(y_test, y_pred_dt) * 100, 3),
        round(accuracy_score(y_test, y_pred_rf) * 100, 3)
    ],
    'Precision': [
        round(precision_score(y_test, y_pred_lr) * 100, 3),
        round(precision_score(y_test, y_pred_dt) * 100, 3),
        round(precision_score(y_test, y_pred_rf) * 100, 3)
    ],
    'Recall': [
        round(recall_score(y_test, y_pred_lr) * 100, 3),
        round(recall_score(y_test, y_pred_dt) * 100, 3),
        round(recall_score(y_test, y_pred_rf) * 100, 3)
    ],
    'F1 Score': [
        round(f1_score(y_test, y_pred_lr) * 100, 3),
        round(f1_score(y_test, y_pred_dt) * 100, 3),
        round(f1_score(y_test, y_pred_rf) * 100, 3)
    ]
})

print(results)
```

```python
import matplotlib.pyplot as plt
import numpy as np

models = ['Logistic Regression', 'Decision Tree', 'Random Forest']

accuracy = [97.461, 99.754, 99.954]
precision = [5.890, 39.286, 89.130]
recall = [91.837, 78.571, 83.673]
f1 = [11.070, 52.381, 86.316]

x = np.arange(len(models))
width = 0.2

plt.figure(figsize=(10,6))

plt.bar(x - 1.5*width, accuracy, width, label='Accuracy')
plt.bar(x - 0.5*width, precision, width, label='Precision')
plt.bar(x + 0.5*width, recall, width, label='Recall')
plt.bar(x + 1.5*width, f1, width, label='F1 Score')

plt.xticks(x, models)
plt.ylabel('Score')
plt.title('Comparison of Machine Learning Models')
plt.legend()
plt.savefig("model_comparison.png", bbox_inches="tight")

plt.show()
```

Random Forest was selected as the final model because it provided the best overall performance across all evaluation metrics. Although Logistic Regression achieved slightly higher Recall, its very low Precision resulted in many false positives. Random Forest offered the best balance between detecting fraudulent transactions and avoiding incorrect fraud predictions, as reflected by its highest F1 Score. Therefore, Random Forest is the most suitable model for the credit card fraud detection task.

5.Analysis


1.Accuracy can be misleading in credit card fraud detection because the dataset is highly imbalanced. Most transactions are legitimate, while only a very small percentage are fraudulent. A model can achieve very high accuracy by simply predicting all transactions as legitimate, but it would fail to detect actual fraud cases. Therefore, accuracy alone does not reflect the model's ability to identify fraudulent transactions effectively.

2.Recall is one of the most important metrics for credit card fraud detection because it measures the proportion of actual fraudulent transactions that are correctly identified by the model. A high recall ensures that fewer fraud cases are missed. Missing a fraudulent transaction can result in financial losses and security risks. However, F1 Score is also important because it balances both Precision and Recall, providing a more comprehensive evaluation of the model's performance.
